### Instanciação do Dash

```python
app = dash.Dash(__name__, external_stylesheets=[dbc.themes.DARKLY])
```

```python
fig = px.choropleth_mapbox(df_states_, locations='estado', color="casosNovos",
                        center={"lat": -16.95, "lon": -47.78},
                        geojson=brazil_states, color_continuous_scale="Redor", opacity=0.4,
                        hover_data={'casosAcumulados': True, "casosNovos": True, "obitosNovos": True, "estado": True})
```

### Layout do Dash

```python
app.layout = dbc.Container(
    dbc.Row([
        dbc.Col([
            dcc.Graph(id="cloropleth-map", figure=fig)
        ])
    ])
)

if __name__ == "__main__":
    app.run_server(debug=True)
```
